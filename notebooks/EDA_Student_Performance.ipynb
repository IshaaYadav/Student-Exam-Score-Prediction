```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
sys.path.append(os.path.abspath(r"C:\Users\ISHA\OneDrive\Documents\GitHub\Student-Exam-Score-Prediction\src"))
from data_preprocessing import load_and_prepare_data


# Set style
sns.set(style='whitegrid')
%matplotlib inline
```

```python
df = load_and_prepare_data()
df.head()
```

```python
df.info()
df.describe().T
df.isnull().sum()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['average_score'], kde=True, color='dodgerblue')
plt.title("Distribution of Average Score")
plt.xlabel("Average Exam Score")
plt.ylabel("Count")
plt.show()
```

```python
features = ['hours_studied', 'previous_score', 'attendance_ratio']
for col in features:
    plt.figure(figsize=(7,4))
    sns.histplot(df[col], kde=True, color='salmon')
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='Blues', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df, diag_kind="kde", corner=True)
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
for feature in features:
    plt.figure(figsize=(7,4))
    sns.scatterplot(x=df[feature], y=df['average_score'], color='mediumseagreen')
    sns.lineplot(x=df[feature], y=df['average_score'], color='black', estimator='mean')
    plt.title(f"{feature} vs Average Score")
    plt.xlabel(feature)
    plt.ylabel("Average Score")
    plt.show()
```

```python
for col in features:
    plt.figure(figsize=(6,4))
    sns.boxplot(y=df[col], color='lightblue')
    plt.title(f"Outlier Check: {col}")
    plt.show()
```
